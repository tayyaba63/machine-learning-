#iloc()  Position-based indexing

Selects rows/columns by their integer position (like array indexing).
Syntax:

df.iloc[row_positions, column_positions]


Example:

image.png


Output:

image.png

```python
df.iloc[34:40]

# rows from position 34 up to (but not including) 40
```

```python
df.loc[(df["Age"] > 25) | (df["Weight"] > 70)]

# return all where age > 25 OR weight>70
```

```python
df.loc[(df["Age"] > 25) | (df["Weight"] > 70), "BMI"]

#  where Age > 25 OR Weight > 70.
# From those rows, return only the BMI column
```

```python
df[(df["Age"] > 40 ) & (df["BMI"] > 30)]


#return all when both condition(age > 40 , BMI > 30) meets
```

#Handling missing values

```python
df.isnull().sum()  # to check null values in dataset
```

```python
df.fillna(method="ffill")
```

###OTHER METHODS TO FILL NULL VALUES :

--> backward fill:


df.fillna(method="bfill")





--> fillna() method

df["col"].fillna(df["col"].mean())       # mean

df["col"].fillna(df["col"].median())     # median

df["col"].fillna(df["col"].mode()[0])     # mode




--> interpolate method


df.interpolate(method="linear")      # default: straight line between values

df.interpolate(method="polynomial", order=2)

df.interpolate(method="time")        # for time series


--> fill with specific value:

df.fillna(0)         # replace with 0

df.fillna("Unknown") # replace with string
