**groupby() function** : The groupby() function in pandas is used to group similar data together and then perform calculations (like sum, average, count, etc.) on each group. It helps in analyzing large datasets by breaking them into smaller, meaningful groups.
give example below :
-> The groupby('Category') groups the rows into "Fruit" and "Vegetable".
-> The sum() function then adds the sales within each category.
-> Now, we can see that the total sales for "Fruit" is 450, and for "Vegetable" is 250.



```python
import pandas as pd

# Creating a DataFrame
data = {'Product': ['Apple', 'Orange', 'Carrot', 'Broccoli', 'Banana'],
        'Category': ['Fruit', 'Fruit', 'Vegetable', 'Vegetable', 'Fruit'],
        'Sales': [100, 150, 120, 130, 200]}

df = pd.DataFrame(data)

print(df)
print("\n")
grouped = df.groupby('Category').sum()  # Group by Category and sum the sales
print("** Apply sum() function** :\n",grouped,"\n\n")
grouped = df.groupby('Category').count() #count() tells us that there are 3 fruit products and 2 vegetable products.
print(" **Apply count() function** : \n",grouped,"\n\n")
grouped = df.groupby('Category')['Sales'].mean()  #The .mean() function calculates the average sales for each category.

print("**Apply mean() function** : \n",grouped,"\n\n")
```

idmax() function :  it  returns the index (row label) of the maximum value in a column.

```python
import pandas as pd

data = {'Product': ['A', 'B', 'C', 'D'],
        'Sales': [100, 250, 300, 150]}

df = pd.DataFrame(data)
print (df,"\n")
# Find the index of the row with the highest sales
max_index = df['Sales'].idxmax()
print("The index of the row with the highest sales :", max_index)
```

agg() function : it is used to Applied multiple functions at once (e.g., sum, mean, min, max).
✅ Works with single or multiple columns.
✅ Works well with groupby() to analyze grouped data.
✅ Supports custom and built-in aggregation functions.


```python
import pandas as pd

# Creating a DataFrame
data = {'Product': ['A', 'B', 'C', 'D'],
        'Sales': [100, 250, 300, 150],
        'Profit': [20, 50, 80, 40]}

df = pd.DataFrame(data)
print (df,"\n")
result = df['Sales'].agg(['sum', 'mean', 'max', 'min']) # Applying multiple functions on 'Sales' column
print("Applying multiple functions on 'Sales' column \n")
print(result,"\n")

result = df.agg({'Sales': ['sum', 'mean'], 'Profit': ['max', 'min']})# Applying different functions to different columns
print("Applying different functions to different columns \n")
print(result,"\n")
```

* using agg()  with custom function  



```python
# Custom function to find range (max - min)
def range_func(x):
    return x.max() - x.min()

# Applying custom function along with built-in functions
result = df.agg({'Sales': ['sum', 'mean', range_func]})
print(result)
```

sort_values() function : This method sorts the DataFrame based on the values in one or more columns (or rows).

**syntax: ** DataFrame.sort_values(by, axis=0, ascending=True, inplace=False)
*by:* The column(s) or row(s) to sort by.

*axis=0*: Sort rows (default), use axis=1 to sort columns

*ascending=True:* Sort in ascending order (use False for descending).

*inplace=False: *Returns a new sorted DataFrame (if True, modifies in place).

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['John', 'Emma', 'Liam', 'Sophia'],
    'Age': [25, 22, 29, 24],
    'Salary': [50000, 70000, 45000, 80000]
})

# Sort by Age (ascending)
df_sorted = df.sort_values(by='Age') # sort only one column
print(df_sorted,"\n")
df_sorted = df.sort_values(by=['Age', 'Salary'], ascending=[True, False]) # sort values of multiple values....First, it sorts by Age in ascending order.If two people have the same Age, it sorts by Salary in descending order.
print(df_sorted , "\n")
df_sorted = df.sort_values(by='Salary', ascending=False)  # sort in descending order
print(df_sorted,"\n")
df_sorted = df.sort_index(by='Age', axis=1) # sort the row
print(df_sorted," \n")



```

**isnull()** : Detects missing values (NaN or None) in a DataFrame or Series.A Boolean mask (True for missing values, False for non-missing values).

**notnull() **: Detects non-missing values (opposite of isnull()).
A Boolean mask (True for non-missing values, False for missing values).

```python
import pandas as pd

# Creating a DataFrame with missing values
data = {'A': [1, 2, None, 4], 'B': [None, 5, 6, 7]}
df = pd.DataFrame(data)

# Detect missing values
print(df.isnull(),"\n")
print(df.notnull(),"\n")
```
